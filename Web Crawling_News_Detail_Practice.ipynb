# 뉴스 상세페이지 조회하기

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url='https://search.naver.com/search.naver'
params={
    'where':'news',
    'query':'펭수'
}
```

```python
resp=requests.get(url,params=params)
resp
```

```python
soup=BeautifulSoup(resp.content)
```

```python
ul_news=soup.find('ul',class_='type01')
li_tags=ul_news.find_all('li',recursive=False)
```

```python
news_list=[]
for li_tag in li_tags:
    news=li_tag.find('a',class_='_sp_each_title')
    url2=news.get('href')
#    print(url2)
    resp2=requests.get(url2)
    soup2=BeautifulSoup(resp2.content)
#    news_list.append(soup2.find('body'))
    ul2_news=soup2.find('body')
    if ul2_news.find('div',class_='article'):
        con_news = ul2_news.find('div',class_='article')
    elif ul2_news.find('article'):
        con_news=ul2_news.find('article')    
    elif ul2_news.find('div', id='article'):
        con_news=ul2_news.find('div', id='article')
    elif ul2_news.find('section'):
        con_news = ul2_news.find('section')
    else:
        con_news=ul2_news
    news_list.append(con_news)
len(news_list)
news_list[1]
```

# 뉴스 상세페이지 

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url='https://search.naver.com/search.naver'
params={
    'where':'news',
    'query':'오늘+날씨'
}
```

```python
resp=requests.get(url,params=params)
resp
```

```python
soup=BeautifulSoup(resp.content)
```

```python
ul_news=soup.find('ul',class_='type01')
news_li_tag=ul_news.find_all('li',recursive=False)
```

```python
my_news_list=[]

for news_tag in news_li_tag :
    a_tag = news_tag.find('a',class_='_sp_each_title')
    title = a_tag.text
    link = a_tag.get('href')
    desc = news_tag.find_all('dd')[1].text
    my_news_list.append({
        'title':title,
        'link':link,
        'desc':desc
    })
    
for idx, my_news in enumerate(my_news_list):
    resp = requests.get(my_news.get('link'))
    soup = BeautifulSoup(resp.content)
    
    my_news_list[idx]['content'] = soup.find('body')
my_news_list
    
```
